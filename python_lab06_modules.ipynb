Чтобы понять пользу от готовых инструментов стандартной библиотеки, давайте попробуем написать свою собственную функцию для возведения в степень.
Функция `power` должна принимать 2 аргумента: `число` и `степень`, в которую нужно возвести данное число, и возвращать результат.
- Дополнительное заданиие: обратите внимание, что нельзя возводить отрицательное число в дробную степень. Сделайте так, чтобы в этом случае функция возвращала 'Ошибка: операция не определена'


```python

def my_power(x, y):
   if x < 0 and type(y) == float:
     return 'Ошибка: операция не определена'
   return x ** y
```

### Задание 3

Напишите функцию-таймер, которая будет принимать время (в секундах), и выводить обратный отсчёт в формате %seconds% ..., пока время не истечёт, а затем выводить 'Время вышло!'. Используйте функцию `sleep` из библиотеки `time`.
- Оцпионально: выводите счётчик оставшегося времени в формате {min}:{sec} (пример: 03:45). Для этого можно воспользоваться целочисленным делением и остатком от деления.

```python
import time
def set_timer(seconds):
    while seconds > 0:
        print(f'осталось {seconds}')
        seconds -= 1
        time.sleep(1)
    print('Время вышло!')

set_timer(4)
```

### Задание 4
Напишите функцию, которая будет выводить количество дней до нового года в формате:
Сегодня %число% . %месяц% . %год%. Значит, до нового года осталось дней: %x

```python
import datetime
today = datetime.datetime.now()
new_year = datetime.datetime(2023,12,31)
delta = new_year - today
delta = delta.days
print(f"Cегодня: {today.strftime('%d.%B.%Y')} Значит, до нового года осталось дней: {delta}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Время (с)!!!', ylabel='напряжение (mV)',
       title='Демонстрационный график')
ax.grid()

fig.savefig("test.png")
plt.show()
```

### Другие примеры графиков

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy

def makeData():
        x = numpy.arange(-10, 10, 0.1)
        y = numpy.arange(-10, 10, 0.1)
        xgrid, ygrid = numpy.meshgrid(x, y)
        zgrid = numpy.sin(xgrid)*numpy.sin(ygrid)/(xgrid*ygrid)
        return xgrid, ygrid, zgrid

x, y, z = makeData()

fig = pylab.figure()
axes = Axes3D(fig)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap=cm.jet)
pylab.show()
```
